## Converting to MFCC

```python
MFCC_Features = []
audio_name = []
MFCC = []
for i in label_names:
    os.chdir(i)
    audio_path = os.listdir()
    for j in audio_path:
        try:
            x , sr = librosa.load(j)
            s = librosa.feature.mfcc(y=x, sr=sr)
            s1 = s.flatten()
            s1 = np.pad(s1,(0,5000-len(s1)),'constant')
            MFCC_Features.append((i,s1))
            MFCC.append(s1)
        except:
            continue
    os.chdir('..')
print('Done')
```

### Creating a Dataframe

```python
l1=np.array(MFCC)
l1.shape
```

```python
df = pd.DataFrame(MFCC_Features, columns = ['Label','MFCC'])
df
```

```python
df1 = pd.DataFrame(l1)
df1
```

```python
frames = [df, df1]
Data = pd.concat(frames,axis=1)
Data
```

### Label Encodings

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Data['Encoded Label']=le.fit_transform(Data['Label'])
Data
```

```python
Data = Data.drop(columns="Label")
Data = Data.drop(columns="MFCC")
```
